```python
import requests
import pandas as pd
from io import StringIO

# URL of the CSV file
csv_url = "https://data.ct.gov/api/views/ucaf-96h6/rows.csv?accessType=DOWNLOAD"

try:
    # Send GET request to fetch the CSV data
    response = requests.get(csv_url)
    response.raise_for_status()  # Check for request errors

    # Read the CSV data into a pandas DataFrame
    csv_data = pd.read_csv(StringIO(response.text))
    
    # Print the first few rows
    print(csv_data.head())

except requests.exceptions.RequestException as e:
    print(f"An error occurred: {e}")
```

```python
csv_data.head()
```

```python
Raw_Data = csv_data
```

```python
Raw_Data.info()
```

```python
import pandas as pd

# Convert 'Week Ending' column to datetime
Raw_Data['Week Ending'] = pd.to_datetime(Raw_Data['Week Ending'], format='%m/%d/%Y')
```

```python
import os
import pandas as pd

# Set the base directory to current working directory
base_dir = os.getcwd()

# Define folder paths
csv_folder = os.path.join(base_dir, 'Clean Raw Data', 'CSV')
xlsx_folder = os.path.join(base_dir, 'Clean Raw Data', 'XLSX')

# Create folders if they do not exist
os.makedirs(csv_folder, exist_ok=True)
os.makedirs(xlsx_folder, exist_ok=True)

# Save as CSV
csv_path = os.path.join(csv_folder, 'Clean_Raw_Data.csv')
Raw_Data.to_csv(csv_path, index=False)

# Save as XLSX
xlsx_path = os.path.join(xlsx_folder, 'Clean_Raw_Data.xlsx')
Raw_Data.to_excel(xlsx_path, index=False)

print(f"CSV saved to: {csv_path}")
print(f"XLSX saved to: {xlsx_path}")
```
