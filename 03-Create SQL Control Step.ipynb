```python
from sqlalchemy import create_engine, text
import pandas as pd

# Datos de conexión
usuario = 'root'
contraseña = '1234'
host = 'localhost'
puerto = 3306
nombre_base = 'Test'

# Crear una conexión sin especificar base de datos
engine_server = create_engine(f'mysql+pymysql://{usuario}:{contraseña}@{host}:{puerto}')

# Crear la base de datos si no existe
with engine_server.connect() as conn:
    conn.execute(text(f"CREATE DATABASE IF NOT EXISTS {nombre_base}"))
    print(f"Base de datos '{nombre_base}' creada o ya existente.")

# Ahora, crear una conexión a la base de datos creada
engine = create_engine(f'mysql+pymysql://{usuario}:{contraseña}@{host}:{puerto}/{nombre_base}')

# Suponiendo que tienes un DataFrame llamado 'Clean_Raw_Data'
# Asegúrate que 'Clean_Raw_Data' esté definido antes de correr esto
# Exporta el DataFrame a la tabla 'Clean_Raw_Data'
Clean_Raw_Data.to_sql('Clean_Raw_Data', con=engine, if_exists='replace', index=False)
```
